Importing the usual suspects and creating some data to plot

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

plt.rcParams["figure.figsize"] = (20,6) # making plot draw full width of notebook 

X = np.random.uniform(-2,5,size=100)
Y = list(map(lambda a: (a-2)**2*(a+1), X))
```

```python
plt.scatter(X,Y)
plt.title = 'y = (x-2)**2 * (x+1)'
plt.ylabel = 'Y values'
plt.xlabel = 'X values'
plt.show()
```
